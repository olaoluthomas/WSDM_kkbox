# Analysis of Categoroical Variables in Member Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import cleaned dataset
members = pd.read_csv('members_cleaned.csv')
members.head()
```

```python
#Check any missing values
members.isnull().any()
```

    1. Gender distribution of member dataset

```python
#Gender
gender = members['gender'].value_counts()
gender.plot(kind='bar',stacked=True, figsize=(15,8))

#plt.xlabel('Smarts')
plt.ylabel('No. of Member')
plt.title("Member's Gender")
plt.show()
```

    2. City distribution of member dataset

```python
#Member's City
city = members['city'].value_counts()
city.plot(kind='bar',stacked=True, figsize=(15,8))

plt.xlabel('City')
plt.ylabel('No. of Member')
plt.title("Member's City")
plt.show()
```

```python
#Member's Registration Method
registered_via = members['registered_via'].value_counts()
registered_via.plot(kind='bar',stacked=True, figsize=(15,8))

plt.xlabel('Registration Method')
plt.ylabel('No. of Member')
plt.title("Member's Registration Method")
plt.show()
```
